```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating input and target variable
x = 10 * np.random.rand(100,)
noise = np.random.randn(100,)
y = 2 * x + noise
```

```python
# Plotting the points
plt.scatter(x, y)
plt.show()
```

```python
alpha = 0.001
m = 100 # Number of data points
```

```python
# Derivative of loss function
def loss_derivative(parameter):
    sum_part = np.sum(np.dot(y - parameter * x, x))
    return - (1 / m) * sum_part
```

```python
# Loss function
def loss(parameter):
    sum_part = np.sum(y - parameter * x)
    return (0.5 / m) * sum_part
```

```python
no_of_iterations = 100
```

```python
# Initializing the parameter
theta = np.random.randn(1)[0]
print(theta)

for i in range(no_of_iterations):
    print(f'Iteration: {i + 1} | Loss = {loss(theta)}')
    if i % 10 == 0:
        plt.figure(figsize = (2, 2))
        plt.scatter(x, y, alpha = 0.3)
        plt.plot(x, theta * x, color = 'red')
        plt.show()
        plt.close()
    theta -= alpha * loss_derivative(theta)
```

```python
print(theta)
```

```python
# Line of best fit
plt.scatter(x, y)
plt.plot(x, theta * x, color = 'red')
plt.show()
```
